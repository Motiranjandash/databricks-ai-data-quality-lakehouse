```python
# Read raw CSV
df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("/Volumes/sales_lakehouse/default/motivolume/product_sales_dataset_final.csv")
)


for col in df.columns:
  df = df.withColumnRenamed(col, col.replace(" ", ""))


# Write to Bronze table
(
    df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("sales_lakehouse.bronze.product_sales_raw")
)

# Validation
spark.table("sales_lakehouse.bronze.product_sales_raw").count()
```
